World report

```r
covid<-read.csv("/content/owid_csv/owid-covid-data.csv", stringsAsFactors = FALSE)
covid$date<-as.Date(covid$date)
head(covid)
dim(covid)
```

Covid Cases in india

```r
install.packages("dplyr")
library(dplyr)
india<-covid%>%filter(location=="India")%>% arrange(date)
head(india)
dim(india)
summary(india)
```

calculate Daily new cases

```r
install.packages("dplyr")
library(dplyr)
india<-covid%>%filter(location=="India")
# Use the existing 'new_cases' column which represents daily new cases
india <- india %>%
  rename(DailyConfirmed = new_cases)
head(india)
```

Visualization through ggplot2

```r
install.packages("ggplot2")
library(ggplot2)
#total confirmed deaths,recovered
ggplot(india,aes(x=date))+
  geom_line(aes(y=total_cases,color="Confirmed"))+
    geom_line(aes(y=total_deaths,color="Deaths"))+
    geom_line(aes(y=(total_cases-total_deaths),color="Recovered")) +
  labs(title="COVID-19 Cases in India", x="Date", y="Number of Cases", color="Legend") +
  theme_minimal()
# Daily new cases
ggplot(india, aes(x=date, y=DailyConfirmed)) +
  geom_bar(stat="identity", fill="purple") +
  labs(title="Daily New Cases in India", x="Date", y="Daily Cases") +
  theme_minimal()
```

Top 10 countries Affected by Covid 19

```r
install.packages("ggplot2")
install.packages("dplyr")
library(ggplot2)
library(dplyr)
top_countries <- covid %>% group_by(location) %>% summarise(TotalConfirmed = max(total_cases, na.rm=TRUE)) %>% arrange(desc(TotalConfirmed)) %>% head(10)
ggplot(top_countries, aes(x=reorder(location, -TotalConfirmed), y=TotalConfirmed)) +
  geom_bar(stat="identity", fill="orange") +
  labs(title="Top 10 Countries by COVID-19 Cases", x="Country", y="Total Confirmed Cases") +
  theme_minimal()
```

Mini report of covid 19 trend analysis in india

```r
peak_day <- india$date[which.max(india$DailyConfirmed)]
peak_cases <- max(india$DailyConfirmed)

report <- paste(
  "Mini Report: COVID-19 Trend Analysis for India\n",
  "Total confirmed cases:", max(india$total_cases, na.rm=TRUE), "\n",
  "Total deaths:", max(india$total_deaths, na.rm=TRUE), "\n",
  "Peak daily new cases:", peak_day, "with", peak_cases, "\n"
)

cat(report)
```
